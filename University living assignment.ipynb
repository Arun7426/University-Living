```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from faker import Faker

# Initialize Faker for generating fake login information
faker = Faker()

# Initialize Chrome webdriver
driver = webdriver.Chrome()

# Navigate to the website
driver.get("https://www.chapter-living.com/")

try:
    # Click on "Book A Room"
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "a[href='/book-a-room']"))).click()

    # Select "Chapter Kings Cross" as the property
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div[data-value='kx']"))).click()

    # Choose the date range from SEP 24 to AUG 25 (51 Days)
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[placeholder='Where']"))).click()
    date_range_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.DayPickerInput input")))
    date_range_input.clear()
    date_range_input.send_keys("SEP 24 to AUG 25 (51 Days)")

    # Filter by room type as Ensuite
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "label[for='roomtype']"))).click()
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div[data-value='Ensuite']"))).click()

    # Iterate through each room detail for BRONZE ENSUITE
    rooms = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".ListingCard__details")))
    for room in rooms:
        room_type = room.find_element_by_css_selector(".ListingCard__name").text
        if "BRONZE ENSUITE" in room_type:
            room.click()
            # Simulate login using Faker library (assuming login is required)
            driver.find_element_by_css_selector("input[name='email']").send_keys(faker.email())
            driver.find_element_by_css_selector("input[name='password']").send_keys(faker.password())
            driver.find_element_by_css_selector("button[type='submit']").click()
            
            # Capture price and availability information
            price = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".Pricing__value"))).text
            availability = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".Pricing__meta"))).text
            
            print("Room Type:", room_type)
            print("Price:", price)
            print("Availability:", availability)
            
            # Go back to the previous page
            driver.back()

finally:
    # Close the browser window
    driver.quit()
```
